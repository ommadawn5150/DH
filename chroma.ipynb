```python
texts = []
import time
for link in get_links:
    res = requests.get(link)
    soup = bs4.BeautifulSoup(res.text, "html.parser")
    # 例として、メインコンテンツが<div class="main-content">にあると仮定
    main_content = soup.find("div", class_="user-area")
    if main_content:
        texts.append(main_content.get_text().replace("\n", "").replace("\u3000", ""))
    time.sleep(0.2)
```
